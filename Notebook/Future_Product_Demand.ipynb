```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:\\Users\\ABHISHEK\\Desktop\\Data Co Supply chaine model\\supply_chain_cleaned_dataset.csv")
```

```python
df.head()
```

```python
df['Order Item Quantity'].unique()
```

```python
# Target variable
y = df["Order Item Quantity"]

# Columns to drop (leakage + IDs)
drop_cols = [
    "Order Item Quantity", "Delivery Status", "Late_delivery_risk", "delay_days",
    "delivery_time", "profit_margin", "order_value_category", "order_value",
    "Order Profit Per Order", "Benefit per order", "Sales", "Order Item Total",
    "Order Item Profit Ratio", "Order Item Product Price", "Order Item Discount Rate",
    "Order Id", "Customer Id", "Order Item Id", "Product Card Id", "Order Item Cardprod Id"
]

X = df.drop(columns=drop_cols, errors="ignore")
```

```python
X
```

```python
y
```

```python

# Encode categorical columns
from sklearn.preprocessing import LabelEncoder
for col in X.select_dtypes(include=["object"]).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_pred = lr_model.predict(X_test)

print("R^2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
future_pred = lr_model.predict(X_test[:10])
print("Predicted demand (first 10 rows):", future_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print("R^2 Score:", score)

```

```python
y_pred = model.predict(X_test[:10])
print("Predicted demand:", y_pred)
```

```python
from sklearn.metrics import mean_squared_error

rf_pred = model.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_pred)
print("Random Forest MSE:", rf_mse)
```

```python
from sklearn.model_selection import cross_val_score

# Random Forest cross-validation
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
scores = cross_val_score(rf_model, X, y, cv=5, scoring='r2')

print("Cross-validation R² scores:", scores)
print("Mean R²:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    scoring='r2'
)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best R² Score:", grid_search.best_score_)
```

```python
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)

print("Tuned Random Forest R²:", r2_score(y_test, y_pred))
print("Tuned Random Forest MSE:", mean_squared_error(y_test, y_pred))
```
